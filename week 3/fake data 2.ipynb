```python
import pandas as pd
import random
from faker import Faker

fd = int(input("How much data required:"))

fake = Faker()

values = {
    "name": [None] * fd,
    "DOB": [None] * fd,
    "phone_no": [None] * fd,
    "email": [None] * fd,
    "address": [None] * fd,
    "kids": [None] * fd,
    "speeding_tickets": [None] * fd,
    "house_sta": [None] * fd,
    "income": [None] * fd,
}

for n in range(fd):
    values["name"][n] = fake.name()
    values["DOB"][n] = fake.year()
    values["phone_no"][n] = fake.phone_number()
    values["email"][n] = fake.email()
    values["address"][n] = fake.address()
    values["kids"][n] = random.randint(0,3)
    values["speeding_tickets"][n] = fake.random_number()
    values["house_sta"][n] = random.choice(["Rent","Own"])
    values["income"][n] = random.randint(10000, 1000000)
      

execute = pd.DataFrame(values)
print('the data is')
execute
```
